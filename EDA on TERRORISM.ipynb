# Selecting the relevant columns

```python
df = terror[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude",
        "location","summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

# Renaming the relevant columns

```python
df.rename(columns = {"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"State",
                    "city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location","summary":"Summary",
                    "attacktype1_txt":"Attack_Type","targtype1_txt":"Target_Type","gname":"Gang_Name","motive":"Motive",
                    "weaptype1_txt":"Weapon_Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add_Notes"
                    },inplace = True)
df.head()
```

# Dimension of Dataset

```python
df.shape
```

The dataset contains 1,81,691 rows and 20 columns.

# Checking for Null Values

```python
df.isnull().sum()
```

There are many null values in rows. So, we are going to fill the null values.

# Filling Null Values

```python
df["Killed"] = df["Killed"].fillna(0)
df["Wounded"] = df["Wounded"].fillna(0)
df["Casualty"] = df["Killed"] + df["Wounded"]
```

```python
df.describe()
```

Observations

# Exploratory Data Analysis

# Visualization

# 1) Year Wise Attacks

Number of attacks in each Year

```python
attacks = df["Year"].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind = "bar", color = "green", figsize = (15,6), fontsize = 12)
plt.title("Attacks Timeline", fontsize = 15)
plt.xlabel("Years", fontsize = 15)
plt.ylabel("No of Attacks", fontsize = 15)
plt.show()
```

1) Most number of attacks (16903) in 2014.

2) Least number of attacks (471) in 1971.

# 2) Total Casualities (Killed + Wounded) in each Year

```python
yc = df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind = "bar", color = "red", figsize = (15,6))
plt.title("Year Wise Casualities", fontsize = 15)
plt.xlabel("Years", fontsize = 15)
plt.ylabel("No of Casualities", fontsize = 15)
plt.show()
```

# 3) Killed in each Year

```python
year_killed = df[["Year","Killed"]].groupby("Year").sum()
year_killed.head()
```

# 4) Wounded in each Region

```python
year_wounded = df[["Year","Wounded"]].groupby("Year").sum()
year_wounded.head()
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,1,1)
ax1 = fig.add_subplot(2,1,2)

# Killed
year_killed.plot(kind = 'bar', color = 'green', figsize = (15,15), ax = ax0)
ax0.set_title("People killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("No.of People Killed")

# Wounded
year_wounded.plot(kind = 'bar', color = 'red', figsize = (15,15), ax = ax1)
ax1.set_title("People wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("No.of People Wounded")

plt.show()
```

# 5) Region wise Attacks

a) Distribution of Terrorist Attacks over Regions from 1970-2017

```python
region = pd.crosstab(df.Year, df.Region)
region.head()
```

```python
region.plot(kind = "area", stacked = False, alpha = 0.5, figsize = (20,10))
plt.title("Region Wise Attacks", fontsize = 20)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("No.of Attacks", fontsize = 20)
plt.show()
```

b) Total Terrorist attacks in each Region from 1970-2017

```python
reg = region.transpose()
reg["Total"] = reg.sum(axis = 1)
ra = reg["Total"].sort_values(ascending = False)
ra
```

```python
ra.plot(kind = "bar", color = "blue", figsize = (15,6))
plt.title("Total no.of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("No.of Attacks")
plt.show()
```

c) Total Casualities (Killed + Wounded) in each Region

```python
rc = df[["Region","Casualty"]].groupby("Region").sum().sort_values(by = "Casualty", ascending = False)
rc
```

```python
rc.plot(kind = "bar", color = "brown", figsize = (15,6))
plt.title("Region Wise Casualities", fontsize = 13)
plt.xlabel("Regions", fontsize = 13)
plt.xticks(fontsize = 12)
plt.ylabel("No.of Casualities", fontsize = 13)
plt.show()
```

d) Killed in each Region

```python
rk = df[["Region","Killed"]].groupby("Region").sum().sort_values(by = "Killed", ascending = False)
rk
```

e) Wounded in each Region

```python
rw = df[["Region","Wounded"]].groupby("Region").sum().sort_values(by = "Wounded", ascending = False)
rw
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

# Killed
rk.plot(kind = 'bar', color = 'green', figsize = (15,6), ax = ax0)
ax0.set_title("People killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("No.of People Killed")

# Wounded
rw.plot(kind = 'bar', color = 'red', figsize = (15,6), ax = ax1)
ax1.set_title("People wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("No.of People Wounded")

plt.show()
```
